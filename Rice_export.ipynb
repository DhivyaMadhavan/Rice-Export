```python
outlier_columns=['IMPORT VALUE FOB','IMPORT VALUE CIF','NET WEIGHT','GROSS WEIGHT','QUANTITY','NUMBER OF PACKAGES','CONTAINER COUNT','TOTAL TEUS']
for i in outlier_columns:
  Q1=final_df[i].quantile(0.25)
  Q3=final_df[i].quantile(0.75)
  IQR=Q3-Q1
  LC=Q1-(1.5*IQR)
  UC=Q3+(1.5*IQR)
  final_df=final_df[(final_df[i]>=LC) & (final_df[i]<=UC)]
```
